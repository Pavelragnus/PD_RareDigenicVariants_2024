```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')
# import dataframe_image as dfi
import re

covar  = pd.read_table('covar_AMP_PD_whole_excl.txt')
covar['Patient'] = covar['FID'] + '_' + covar['IID'] +'.csv'
covar = covar.astype({'Patient':str})
covar = covar.set_index('Patient')
path = os.getcwd() 
l = os.listdir(path)
ls = [i for i in l if i.endswith('csv')]
dfs = pd.DataFrame()
for i in ls:
    d = pd.read_csv(i)
    print(f'{i} : {len(d)}')
    dfs = pd.concat([dfs, d])
dfs = dfs.rename(columns = {'Unnamed: 0': 'Patient'})
```

```python
df1 = dfs[['HGVSc', 'ClinVar_CLNDN']]
df1 = df1.rename(columns={"ClinVar_CLNDN": "ClinVar_CLNDN 1"})
df1 = df1.set_index('HGVSc')
df1 = df1.drop_duplicates()
```

```python
df2 = dfs[['HGVSc', 'ClinVar_CLNDN']]
df2 = df2.rename(columns={"ClinVar_CLNDN": "ClinVar_CLNDN 2"})
df2 = df2.set_index('HGVSc')
df2 = df2.drop_duplicates()
```

```python
d = pd.read_csv('CV_correct/digens_with_MAF_wholeexcl.csv')
```

```python

d[['Gene1', 'Mutation1 only']] = d['Mutation1 only'].str.split(' ', expand=True)
d[['Gene2', 'Mutation2 only']] = d['Mutation2 only'].str.split(' ', expand=True)
d = d.drop(['Gene1', 'Gene2'], axis = 1)
df1['Mutation1 only'] = df1.index
d = d.set_index(["Mutation1 only"])
df1 = df1.set_index(["Mutation1 only"])
res = d.join(df1, on = 'Mutation1 only', how ="left").reset_index().drop(['Unnamed: 0'], axis = 1)
```

```python

df2['Mutation2 only'] = df2.index
d = d.set_index(["Mutation2 only"])
df2 = df2.set_index(["Mutation2 only"])
res = res.join(df2, on = 'Mutation2 only', how ="left").reset_index().drop(['index'], axis = 1)
```
