```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('GBA1/binary_result.csv')
```

```python
result = df
```

```python
result.columns = result.columns.str.replace(',', '_')
```

```python
result.columns = result.columns.str.replace('/', '_')
```

```python
result.columns = result.columns.str.replace(':', '_')
```

```python
result.columns = result.columns.str.replace('>', '_')
```

```python
result.columns = result.columns.str.replace('.', '_')
```

```python
result.columns = result.columns.str.replace('-', '_')
```

```python
result.columns = result.columns.str.replace('*', '')
```

```python
factors = df.columns[13::]
```

```python
result.columns
```

```python
formula = 'Status ~ Age + Sex'
formula = formula + f' + {factors[0]}'
model = smf.glm(formula = formula, data=result, family=sm.families.Binomial())
glm = model.fit()
y = (glm.summary2().tables[1])
for i in factors[1::]:
    formula = 'Status ~ Age + Sex'
    formula = formula + f' + {i}'
    model = smf.glm(formula = formula, data=result, family=sm.families.Binomial())
    glm = model.fit()
    r = (glm.summary2().tables[1])
    y = pd.concat([y, r.iloc[[-1]]])
```

```python
y.to_csv('GBA1/glm_summary.csv')
```
